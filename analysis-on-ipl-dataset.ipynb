```python
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
plt.figure(figsize=(9,7))
plt.bar(list(batting_first['winner'].value_counts()[0:4].keys()),
        list(batting_first['winner'].value_counts()[0:4]),
        color=('green','Yellow','Red','Blue','orange'))
plt.show()
```
